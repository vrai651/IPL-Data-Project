# Chennai Super Kings (M. A. Chidambaram Stadium records)
M. A. Chidambaram Stadium matchs

```python
Chennai = matchinfodata[matchinfodata['city'] == 'Chennai']
Chennai
```

#### M. A. Chidambaram Stadium matchs win

```python
Chennaiwin = Chennai[Chennai['winner'] == 'Chennai Super Kings']
Chennaiwin
```

#### M. A. Chidambaram Stadium matchs win count

```python
Chennaiwincount = len(Chennaiwin)
Chennaiwincount
```

##### M. A. Chidambaram Stadium matchs lost

```python
Chennailost = Chennai[Chennai['winner'] != 'Chennai Super Kings']
Chennailost
```

#### M. A. Chidambaram Stadium matchs lost count

```python
Chennailostcount = len(Chennailost)
Chennailostcount
```

### summary of win and lost

```python
print("Chennai Super Kings Wins in M. A. Chidambaram Stadium:", Chennaiwincount)
print("Chennai Super Kings Losses in M. A. Chidambaram Stadium:", Chennailostcount)
```

### Average number of matches won by CSK in Chennai

```python
total_matches = len(Chennai)
total_wins = len(Chennai[Chennai['winner'] == 'Chennai Super Kings'])
average_win = total_wins / total_matches

print("Average number of matches won by Chennai Super Kings  in Chennai:", average_win)
```

### Average number of matches lost by CSK in Chennai

```python
total_matches = len(Chennai)
total_losses = len(Chennai[Chennai['winner'] != 'Chennai Super Kings'])

average_loss = total_losses / total_matches

print("Average number of matches lost by Chennai Super Kings  in Chennai:", average_loss)
```
